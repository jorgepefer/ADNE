### HISTOGRAMAS

```python
import os

def generar_histograma_aleatorio(carpeta_destino ):
    distribuciones = ['normal', 'uniform', 'exponential', 'gamma', 'lognormal', 'weibull', 'chi_squared']
    distribucion_elegida = random.choice(distribuciones)
    
    if distribucion_elegida == 'normal':
        media = random.uniform(-5, 5)
        desviacion = random.uniform(0.5, 3)
        datos = np.random.normal(media, desviacion, 1000)
        folder_path = 'normal_histograms'
    elif distribucion_elegida == 'uniform':
        inicio = random.uniform(-10, 0)
        fin = random.uniform(0, 10)
        datos = np.random.uniform(inicio, fin, 1000)
        folder_path = 'uniform_histograms'
    elif distribucion_elegida == 'exponential':
        escala = random.uniform(0.5, 2)
        datos = np.random.exponential(escala, 1000)
        folder_path = 'exponential_histograms'
    elif distribucion_elegida == 'gamma':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.gamma(forma, escala, 1000)
        folder_path = 'gamma_histograms'
    elif distribucion_elegida == 'lognormal':
        media = random.uniform(-2, 2)
        sigma = random.uniform(0.1, 1)
        datos = np.random.lognormal(media, sigma, 1000)
        folder_path = 'lognormal_histograms'
    elif distribucion_elegida == 'weibull':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.weibull(forma, 1000) * escala
        folder_path = 'weibull_histograms'
    elif distribucion_elegida == 'chi_squared':
        grados_libertad = random.randint(1, 10)
        datos = np.random.chisquare(grados_libertad, 1000)
        folder_path = 'chi_squared_histograms'
    num_bins = random.randint(5, 50)
    
    color = np.random.rand(3,)
    
    plt.hist(datos, bins=num_bins, color=color, edgecolor='black')
    plt.title(f'Histograma Aleatorio ({distribucion_elegida.capitalize()})')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    

    carpeta_destino=os.path.join(carpeta_destino,folder_path)
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
    nombre_archivo = f'histograma_{random.randint(1, 1000000)}.png'
    nombre_archivo = os.path.join(carpeta_destino, nombre_archivo)
    borrar_archivo(nombre_archivo)  # Solo si existe   
    plt.savefig(nombre_archivo)
    plt.close()
    
    return nombre_archivo
```

```python

def descargar_histogramas(num_descargas, carpeta_destino):

    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
   
    
    for i in range(num_descargas):
        nombre_archivo = generar_histograma_aleatorio(carpeta_destino)
        print(f"Descargado {nombre_archivo}")


```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import os

def generar_boxplot_aleatorio(carpeta_destino):
    distribuciones = ['normal', 'uniform', 'exponential', 'gamma', 'lognormal', 'weibull', 'chi_squared', 'random']
    distribucion_elegida = random.choice(distribuciones)
    
    if distribucion_elegida == 'normal':
        media = random.uniform(-5, 5)
        desviacion = random.uniform(0.5, 3)
        datos = np.random.normal(media, desviacion, 1000)
        folder_path = 'normal_boxplots'
    elif distribucion_elegida == 'uniform':
        inicio = random.uniform(-10, 0)
        fin = random.uniform(0, 10)
        datos = np.random.uniform(inicio, fin, 1000)
        folder_path = 'uniform_boxplots'
    elif distribucion_elegida == 'exponential':
        escala = random.uniform(0.5, 2)
        datos = np.random.exponential(escala, 1000)
        folder_path = 'exponential_boxplots'
    elif distribucion_elegida == 'gamma':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.gamma(forma, escala, 1000)
        folder_path = 'gamma_boxplots'
    elif distribucion_elegida == 'lognormal':
        media = random.uniform(-2, 2)
        sigma = random.uniform(0.1, 1)
        datos = np.random.lognormal(media, sigma, 1000)
        folder_path = 'lognormal_boxplots'
    elif distribucion_elegida == 'weibull':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.weibull(forma, 1000) * escala
        folder_path = 'weibull_boxplots'
    elif distribucion_elegida == 'chi_squared':
        grados_libertad = random.randint(1, 10)
        datos = np.random.chisquare(grados_libertad, 1000)
        folder_path = 'chi_squared_boxplots'
    elif distribucion_elegida == 'random':
        datos = np.random.rand(1000)  # Completely random data
        folder_path = 'random_boxplots'
    else:
        raise ValueError("Invalid distribution")
    
    color = np.random.rand(3,)
    
    
    # Define los colores para cada parte del boxplot
    boxplot_colors = {
        'boxes': color,
        'whiskers': color,
        'medians': color,
        'caps': color,
        'fliers': color,
        'means': color
    }
    
    # Crea el boxplot con colores
    plt.boxplot(datos, patch_artist=True, boxprops=dict(facecolor=color), capprops=dict(color=color),
                whiskerprops=dict(color=color), flierprops=dict(color=color, markeredgecolor=color),
                medianprops=dict(color=color), meanprops=dict(color=color))
    plt.boxplot(datos)
    plt.title(f'Boxplot Aleatorio ({distribucion_elegida.capitalize()})')
    plt.ylabel('Valor')
    
    carpeta_destino=os.path.join(carpeta_destino,folder_path)
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
    nombre_archivo = f'boxplot_{random.randint(1, 1000000)}.png'
    nombre_archivo=os.path.join(carpeta_destino,nombre_archivo)
    borrar_archivo(nombre_archivo)  # Solo si existe
    plt.savefig(nombre_archivo)
    plt.close()
    

    return nombre_archivo
```

```python
def descargar_boxplots(num_descargas, carpeta_destino):

    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)

    
    for i in range(num_descargas):
        nombre_archivo = generar_boxplot_aleatorio(carpeta_destino)
        
        ruta_destino = os.path.join(carpeta_destino, nombre_archivo)
        print(f"Descargado {ruta_destino}")
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import os
from collections import Counter

def absolute_frequency(numbers):
    freq_counter = Counter(numbers)
    freq_dict = dict(freq_counter)
    return freq_dict

def generar_barplot_aleatorio_cat(carpeta_destino):
    # Generate random data for the categorical variable
    num_categories = random.randint(2, 7)  # Random number of categories between 3 and 7
    categories = [f'Category_{i}' for i in range(1, num_categories + 1)]  # Random category labels
    
    # Choose a random distribution for the categorical variable
    distributions = ['bernoulli', 'binomial', 'poisson', 'geometric', 'random']
    distribution = random.choice(distributions)
    
    # Generate random samples based on the chosen distribution
    if distribution == 'bernoulli':
        p = random.uniform(0.1, 0.9)
        samples = np.random.choice([0, 1], size=random.randint(10, 50), p=[1-p, p])
        folder_path = 'bernoulli_barplots'
    elif distribution == 'binomial':
        n = random.randint(5, 20)
        p = random.uniform(0.1, 0.9)
        samples = np.random.binomial(n, p, size=random.randint(10, 50))
        folder_path = 'binomial_barplots'
    elif distribution == 'poisson':
        lambda_ = random.uniform(1, 10)
        samples = np.random.poisson(lambda_, size=random.randint(10, 50))
        folder_path = 'poisson_barplots'
    elif distribution == 'geometric':
        p = random.uniform(0.1, 0.9)
        samples = np.random.geometric(p, size=random.randint(10, 50))
        folder_path = 'geometric_barplots'
    elif distribution == 'random':
        samples = np.random.randint(1, 50, size=random.randint(10, 50))
        folder_path = 'random_barplots'
    else:
        raise ValueError("Invalid categorical distribution")
    
    # Calculate absolute frequency of each sample
    freq_dict = absolute_frequency(samples)
    
    # Generate random colors for the bars
    bar_colors = [tuple(np.random.rand(3)) for _ in range(len(freq_dict))]
    
    # Create bar plot
    plt.bar(freq_dict.keys(), freq_dict.values(), color=bar_colors)
    plt.title('Barplot Aleatorio (Categorical)')
    plt.xlabel('Categories')
    plt.ylabel('Frequency')
    
    # Set x-tick labels to only show categories in freq_dict
    plt.xticks(list(freq_dict.keys()))
    
    # Save the plot
    carpeta_destino = os.path.join(carpeta_destino, folder_path)
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
    nombre_archivo = f'barplot_cat_{random.randint(1, 1000000)}.png'
    nombre_archivo = os.path.join(carpeta_destino, nombre_archivo)
    plt.savefig(nombre_archivo)
    plt.close()
    
    return nombre_archivo
```

```python
def descargar_barplots(num_descargas, carpeta_destino):

    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)    

    for i in range(num_descargas):
        nombre_archivo=generar_barplot_aleatorio_cat(carpeta_destino)
        ruta_destino = os.path.join(carpeta_destino, nombre_archivo)
        print(f"Descargado {ruta_destino}")
```

```python
 
import random
import matplotlib.pyplot as plt
import os
from collections import Counter

def absolute_frequency(numbers):
    freq_counter = Counter(numbers)
    freq_dict = dict(freq_counter)
    return freq_dict

def generar_pieplot_aleatorio_cat(carpeta_destino):
    # Generate random data for the categorical variable
    num_categories = random.randint(2, 7)  # Random number of categories between 3 and 7
    categories = [f'Category_{i}' for i in range(1, num_categories + 1)]  # Random category labels
    
    # Choose a random distribution for the categorical variable
    distributions = ['bernoulli', 'binomial', 'poisson', 'geometric', 'random']
    distribution = random.choice(distributions)
    
    # Generate random samples based on the chosen distribution
    if distribution == 'bernoulli':
        p = random.uniform(0.1, 0.9)
        samples = np.random.choice([0, 1], size=random.randint(10, 50), p=[1-p, p])
        folder_path = 'bernoulli_pieplots'
    elif distribution == 'binomial':
        n = random.randint(5, 20)
        p = random.uniform(0.1, 0.9)
        samples = np.random.binomial(n, p, size=random.randint(10, 50))
        folder_path = 'binomial_pieplots'
    elif distribution == 'poisson':
        lambda_ = random.uniform(1, 10)
        samples = np.random.poisson(lambda_, size=random.randint(10, 50))
        folder_path = 'poisson_pieplots'
    elif distribution == 'geometric':
        p = random.uniform(0.1, 0.9)
        samples = np.random.geometric(p, size=random.randint(10, 50))
        folder_path = 'geometric_pieplots'
    elif distribution == 'random':
        samples = np.random.randint(1, 50, size=random.randint(10, 50))
        folder_path = 'random_pieplots'
    else:
        raise ValueError("Invalid categorical distribution")
    
    # Calculate absolute frequency of each sample
    freq_dict = absolute_frequency(samples)
    
    # Generate random colors for the slices
    pie_colors = [tuple(np.random.rand(3)) for _ in range(len(freq_dict))]
    
    # Create pie plot
    plt.pie(freq_dict.values(), labels=freq_dict.keys(), colors=pie_colors, startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Pieplot Aleatorio (Categorical)')
    
    # Save the plot
    carpeta_destino = os.path.join(carpeta_destino, folder_path)
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
    nombre_archivo = f'pieplot_cat_{random.randint(1, 1000000)}.png'
    nombre_archivo = os.path.join(carpeta_destino, nombre_archivo)
    plt.savefig(nombre_archivo)
    plt.close()
    
    return nombre_archivo
```
