```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

def generar_histograma_aleatorio():

    distribuciones = ['normal', 'uniform', 'exponential', 'gamma', 'lognormal', 'poisson', 'bernoulli', 'binomial', 'weibull', 'chi_squared']
    distribucion_elegida = random.choice(distribuciones)
    
    if distribucion_elegida == 'normal':
        media = random.uniform(-5, 5)
        desviacion = random.uniform(0.5, 3)
        datos = np.random.normal(media, desviacion, 1000)
    elif distribucion_elegida == 'uniform':
        inicio = random.uniform(-10, 0)
        fin = random.uniform(0, 10)
        datos = np.random.uniform(inicio, fin, 1000)
    elif distribucion_elegida == 'exponential':
        escala = random.uniform(0.5, 2)
        datos = np.random.exponential(escala, 1000)
    elif distribucion_elegida == 'gamma':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.gamma(forma, escala, 1000)
    elif distribucion_elegida == 'lognormal':
        media = random.uniform(-2, 2)
        sigma = random.uniform(0.1, 1)
        datos = np.random.lognormal(media, sigma, 1000)
    elif distribucion_elegida == 'poisson':
        lambda_ = random.uniform(1, 10)
        datos = np.random.poisson(lambda_, 1000)
    elif distribucion_elegida == 'bernoulli':
        p = random.uniform(0.1, 0.9)
        datos = np.random.choice([0, 1], size=1000, p=[1-p, p])
    elif distribucion_elegida == 'binomial':
        n = random.randint(5, 20)
        p = random.uniform(0.1, 0.9)
        datos = np.random.binomial(n, p, 1000)
    elif distribucion_elegida == 'weibull':
        forma = random.uniform(1, 5)
        escala = random.uniform(0.5, 2)
        datos = np.random.weibull(forma, 1000) * escala
    elif distribucion_elegida == 'chi_squared':
        grados_libertad = random.randint(1, 10)
        datos = np.random.chisquare(grados_libertad, 1000)
    
    num_bins = random.randint(10, 50)
    
    color = np.random.rand(3,)
    
    plt.hist(datos, bins=num_bins, color=color, edgecolor='black')
    plt.title(f'Histograma Aleatorio ({distribucion_elegida.capitalize()})')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    
    nombre_archivo = f'histograma_{random.randint(1, 1000000)}.png'
    plt.savefig(nombre_archivo)
    plt.close()
    
    return nombre_archivo

def descargar_histogramas(num_descargas, carpeta_destino):

    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
    
    for i in range(num_descargas):
        nombre_archivo = generar_histograma_aleatorio()
        ruta_destino = os.path.join(carpeta_destino, nombre_archivo)
        os.rename(nombre_archivo, ruta_destino)
        print(f"Descargado {ruta_destino}")

num_descargas = 100
carpeta_destino = 'histogramas'

descargar_histogramas(num_descargas, carpeta_destino)
```
